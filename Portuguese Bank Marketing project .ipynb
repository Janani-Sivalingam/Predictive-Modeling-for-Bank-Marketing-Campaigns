### Insight


1. Marital Status Impact
Just being married doesn't necessarily mean someone will say yes to a term deposit. While married folks do say yes more often than single or divorced people, they're also more likely to say no overall.

2. Default History
People who've never missed payments (no defaults) are more open to term deposits than those with past defaults or unknown history. But don't get too excited - they're still more likely to say no than yes.

3. Housing Loans
Having a housing loan might make someone slightly more interested in a term deposit. But here's the catch - these same people are still more likely to turn down the offer than accept it.

4. Personal Loans
Interestingly, people without any loans are more receptive to term deposits than those juggling existing loans. Though, like the others, they still tend to say no more often than yes.

5. Contact Method
Cell phones seem to be the way to go! People contacted on their cell phones are more likely to consider a term deposit than those reached on landlines. But again, even with cell phones, you'll hear more nos than yeses.

6. Timing Matters
The good old workweek (Monday to Friday) seems to be the sweet spot for getting people interested. But don't expect miracles - even during these prime days, most people still pass on the offer.

7. Past Experience
Most customers are new contacts (marked as 'nonexistent' previous outcome). These fresh faces are more likely to consider the offer compared to those who've said no before or even those who've said yes in the past. But - you guessed it - they're still morstill an uphill battle!

## Hyperparameter Tuning for Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()

from sklearn.model_selection import RandomizedSearchCV

# Define model
log_reg = LogisticRegression()

# Define hyperparameter grid
param_grid = {
    'penalty': ['l1', 'l2', 'elasticnet', None],
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'saga'],
    'max_iter': [100, 200, 300]
}

# Randomized search
random_search = RandomizedSearchCV(log_reg, param_grid, cv=5, scoring='accuracy', n_iter=10, n_jobs=-1, random_state=42)
random_search.fit(X_train, y_train)

# Best parameters
print("Best Hyperparameters:", random_search.best_params_)
```

## Hyperparameter for Random Forest¶
